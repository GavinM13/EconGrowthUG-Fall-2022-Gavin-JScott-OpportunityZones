```python
subsamp_df = pd.read_csv("main_data.csv.zip")
```

```python
counties_df = pd.read_excel("ipums_usa_identified_counties.xlsx", skiprows=1)
```

```python
counties_df = counties_df.rename(
    columns={"Unnamed: 0": "State", "Unnamed: 1": "County"}
)
counties_df = counties_df[
    ["State", "County", "STATEFIP", "COUNTYICP", "COUNTYFIP"]
].copy()
counties_df[counties_df["COUNTYICP"] == 730]
```

```python
oz_df = pd.read_excel("designated-qozs.12.14.18.xlsx", skiprows=4)
oz_df = oz_df.rename(columns={"Click arrow to filter state\n\nState": "State"})
```

```python
oz_df.head()
```

```python
merge1_df = pd.merge(
    subsamp_df,
    counties_df,
    left_on=["STATEFIP", "COUNTYICP"],
    right_on=["STATEFIP", "COUNTYICP"],
)
```

```python
merge2_df = pd.merge(
    merge1_df, oz_df, how="left", on=["State", "County"], indicator=True
)
```

```python
merge2_df.head()
```

```python
main_df = merge2_df[
    (merge2_df["HHINCOME"] != 9999999)
    & (merge2_df["HHINCOME"] > 0)
    & (merge2_df["_merge"] != "right_only")
].copy()
```

```python
# OZ or not
main_df["OZ"] = main_df["_merge"] == "both"

# Treatment period or not
main_df["treatment"] = main_df["YEAR"] > 2017

# Interaction
main_df["OZxTreatment"] = main_df["OZ"] & main_df["treatment"]

# Control for inflation
# main_df["Inf_Cont_HHINCOME"]...
```

# Opportunity Zones

## The question: Do opportunity zones cause an increase in income for the people living inside of an opportunity zone?

## What is an Opportunity Zone?

Opportunity Zones are an economic development tool that allows people to invest in distressed areas in the United States.

Their purpose is to spur economic growth and job creation in low-income communities while providing tax benefits to investors.

## How do Opportunity Zones work?

According to the Economic Innovation Group, "Investors can choose to roll capital gains over into qualified Opportunity Funds, which in turn channel patient capital into qualifying equity investments in Opportunity Zones for at least a decade in exchange for capital gains tax reductions and possible exemptions. This new source of risk capital will seed new startups, accelerate business expansions, create jobs, increase and improve housing options, and revitalize the built environment in distressed communities across the country"

## Difference in Difference (DID) Explanation

Difference in Difference is a type of regression in which there is a treated and untreated group, in our case these treated and untreated groups or people in opportunity zones and not opportunity zones. We also consider a period in time where treatment starts and treatment does not start, basically when the policy of opportunity zones takes effect in which we compare before and after. Through this type of regression we are also able to control for indicators of income so we can treat all responses the same, (creating a ceteris paribus), and minimize the error in our work.  

## What data indicators did we use?

For our data indicators total household income was our dependent variable and our geographic variables and other economic indicators were control variables.

$$\ln(\text{Mean_HH_Income}) = \alpha\cdot\text{OZ} + \beta\cdot\text{T} + \gamma\cdot(\text{OZ}\cdot\text{T}) + C_1\cdot\text{FOODSTMP} + C_2\cdot\text{VALUEH}$$

```python
This is the DID equation we constructed in which OZ * T is our DID variable
```

```python
mean_hhincome_oz_grouped_df["ln_mean_hhincome"] = np.log(
    mean_hhincome_oz_grouped_df["HHINCOME"]
)

linear_model = ols(
    "ln_mean_hhincome ~ OZ + treatment + OZxTreatment + FOODSTMP + VALUEH", data=mean_hhincome_oz_grouped_df
).fit()
```

```python
quantile_oz_grouped_df_dict = {}

for prcnt in [0.10, 0.25, 0.50, 0.75, 0.90]:
    quantile_oz_grouped_df_dict[prcnt] = (
        main_df.groupby(by=["OZ", "YEAR"]).quantile(prcnt).reset_index()
    )
    quantile_oz_grouped_df_dict[prcnt]["ln_mean_hhincome"] = np.log(
        quantile_oz_grouped_df_dict[prcnt]["HHINCOME"]
    )
    linear_model = ols(
        "ln_mean_hhincome ~ OZ + treatment + OZxTreatment + FOODSTMP + VALUEH",
        data=quantile_oz_grouped_df_dict[prcnt],
    ).fit()
    print(f"Results for {prcnt}-percentile income")
    print(linear_model.summary())
    print()
```

```python
for prcnt in [0.10, 0.25, 0.50, 0.75, 0.90]:
    linear_model = ols(
        "ln_mean_hhincome ~ OZ + treatment + OZxTreatment + FOODSTMP + VALUEH",
        data=quantile_oz_grouped_df_dict[prcnt],
    ).fit()
    print(f"Results for {prcnt}-percentile income")
    print(linear_model.summary())
    print()
```

```python
# Without grouping by year
quantile_oz_grouped_df_dict_2 = {}

percentiles = [0.0, 0.10, 0.25, 0.50, 0.75, 0.90, 1.0]

for prcnt in percentiles:
    quantile_oz_grouped_df_dict_2[prcnt] = (
        main_df.groupby(by=["OZ"])["HHINCOME"].quantile(prcnt)
    ).reset_index()
   
```

```python
for i in range(len(percentiles) - 1):
    tmp_df = main_df[
        (
            (
                (main_df["OZ"] == True)
                & (
                    main_df["HHINCOME"]
                    > quantile_oz_grouped_df_dict_2[percentiles[i]].iloc[0]["HHINCOME"]
                )
            )
            | (
                (main_df["OZ"] == False)
                & (
                    main_df["HHINCOME"]
                    > quantile_oz_grouped_df_dict_2[percentiles[i]].iloc[1]["HHINCOME"]
                )
            )
        )
        & (
            (
                (main_df["OZ"] == True)
                & (
                    main_df["HHINCOME"]
                    < quantile_oz_grouped_df_dict_2[percentiles[i + 1]].iloc[0][
                        "HHINCOME"
                    ]
                )
            )
            | (
                (main_df["OZ"] == False)
                & (
                    main_df["HHINCOME"]
                    < quantile_oz_grouped_df_dict_2[percentiles[i + 1]].iloc[1][
                        "HHINCOME"
                    ]
                )
            )
        )
    ]
    tmp_df["ln_hhincome"] = np.log(tmp_df["HHINCOME"])
    linear_model = ols(
        "ln_hhincome ~ OZ + treatment + OZxTreatment + FOODSTMP + VALUEH",
        data=tmp_df,
    ).fit()
    print(f"Results for {percentiles[i + 1]}-percentile income")
    print(linear_model.summary())
    print()
```

## Data Limitations

Limitations to our data is that we were only able to select a million of our respondents at a time, for the reason that if we had selected all 20 million our computer would have major errors and not be able to run the regression. Furthermore, other limitations to our study consist of not being able to directly match each respondent to an opportunity zones, because they were matched by census tract number and that was private information on IPUMS. Lastly we only have two indicators of wealth, Food Stamps and Housevalue, because they were the only two that had information in all of the years that we were looking for. 

## Conclusions

In conclusion it was always going to be difficult to find a direct statistical answer to this question due to the data limitations, and in fact we cannot confirm or deny that there is causation between opportunity zones, and an increase or even a decrease in median household income for people living in them. Furthermore, due to the nature of the timeline for opportunity zone investment, it is most likely too early to tell if opportunity zones work or not, and it may be better to wait and see until 2027 when the tax write-offs take place.
